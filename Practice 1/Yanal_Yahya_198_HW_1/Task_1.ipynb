## Задание 1

Вам поручено разработать онлайн-аукцион. Он позволяет продавцам продавать свои товары с помощью аукциона. Покупатели делают ставки. Выигрывает последняя самая высокая ставка. После закрытия аукциона победитель оплачивает товар с помощью кредитной карты. Продавец отвечает за доставку товара покупателю.

* Предложите список функциональных требований для проекта.
* Определите роли пользователей и действия для каждой роли.
* Определите объекты, о которых будут храниться данные.
* Определите связи между объектами для хранения данных.
* Нарисуйте схему объектной модели (используя любые обозначения, которые вам удобны).

### Соглашение о приеме работ

* В папке с названием задания, создается папка с фамилией студента, в которой он работает
* Текстовы файлы оформляются в формате .md . Если нет возможности - pdf, word, etc.
* Изображения лежат рядом, в .md указываем ссылку
* Срок сдачи - до 00:00 11.09.2021. 
* Комиты можно писать на русском языке

#1. Функциональные требования

* Авторизация
 * Аутентификация должна выполняться на уровне пользователя.
 * Для всех функций онлайн-аукциона потребуется аутентификация.

* Помощь
 * Пользователь должен иметь возможность получить доступ к контекстно-зависимой справке из любого места.

 * У пользователя должны быть основные инструкции, доступные на текущей странице, относящиеся к теме страницы.

 * Пользователи смогут просматривать все свои текущие транзакции (запросы на товары, покупку и продажу) одновременно.

* Продавать
 * Пользователь должен иметь возможность выставить товар на продажу.

 * Пользователь должен иметь возможность устанавливать начальную цену предложения.

 * У пользователя должна быть возможность закрыть аукцион через 24 часа после запроса о закрытии аукциона с невыполненными ставками.

 * Пользователь должен иметь возможность немедленно закрыть аукцион, если никто не сделал ставку на этот предмет.

 * Пользователь должен иметь возможность установить продолжительность аукциона по умолчанию в любой момент от трех до 30 дней.

 * Пользователь должен иметь возможность размещать текстовое описание объекта.

 * Пользователь должен иметь возможность публиковать изображения товара.

 * Пользователь должен иметь возможность установить минимальную цену «Go» [цена, по которой товар может быть продан; если аукцион не соответствует этому пункту, результат аукциона не является обязательным], который будет оставаться конфиденциальным. 

* Торги по позициям
 * Пользователь должен иметь возможность делать ставки на товары, доступные для продажи.

 * Все заявки являются обязательными.
   * Невыигравшие заявки не являются обязательными.

   * Выигравшие ставки могут быть отклонены по соглашению обеих сторон.

 * Будут доступны различные типы / стили ставок.
   * Пользователь может установить максимальную ставку, выше которой он не будет делать ставки. Система автоматически увеличит их ставку в соответствии с их стилем ставки до максимального значения. 
 * Победитель оплачивает товар с помощью кредитной карты

* Поиск в базе данных
 * Пользователь должен иметь возможность искать доступные элементы по категории.

 * Пользователь должен иметь возможность искать доступные элементы по заголовку.

 * Пользователь должен иметь возможность искать доступные товары по продавцу.

 * Пользователь должен иметь возможность искать запрашиваемые элементы по категории.

 * Пользователь должен иметь возможность искать запрашиваемые элементы по заголовку.

 * Пользователь должен иметь возможность искать запрошенные товары по данным продавца.

 * Пользователь должен иметь возможность искать доступные книги по категории.

 * Пользователь должен иметь возможность искать доступные книги по названию.

 * Пользователь должен иметь возможность искать доступные книги по продавцу.

 * Пользователь должен иметь возможность искать доступные книги по автору.
